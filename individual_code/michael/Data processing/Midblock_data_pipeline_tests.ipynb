```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv("Midblock_Traffic_Counts.csv")
```

```python
df.head()
```

```python
df2 = pd.read_csv("processing_results3.csv")
df2.head()
```

```python
test = df2['0'].str.split(" & ", n=1, expand = True)
test.columns = ['Street', 'Street From']
test = test.join(df2['lat'])
test = test.join(df2['long'])
test
```

```python
new_df = pd.merge(df, test,  how='left', left_on=['Street','Street From'], right_on = ['Street','Street From'])
new_df.head()
```

```python
print(df.shape)
print(new_df.shape)
# Not sure why the row values INCREASED... better off remaking Geoprocessing.ipynb when we figure out our final data pipeline
```

Using google maps, get a rough idea of the range of lat/long values around St. Vital mall:

49.819851, -97.123511
49.836164, -97.104169

```python
st_vital = new_df.loc[(new_df['lat'] >= 49.819851) & (new_df['lat'] <= 49.836164) &
           (new_df['long'] >= -97.123511) & (new_df['long'] <= -97.104169)]
```
